```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/housing.csv")
```

```python
# See top records
print(df.head())
```

```python
# Summary statistics
print(df.describe())
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Correlation heatmap
sns.heatmap(df.corr(), annot=True)
plt.title("Feature Correlation")
plt.show()
```

```python
# Pairplot to visualize distributions
sns.pairplot(df)
plt.show()
```
